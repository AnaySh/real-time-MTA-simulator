```python
import pandas as pd

# Number of people who got on at 8 Av
line = "L"
origin_complex_id = "618"

# Initialize variables
num_stations = complexes.get_number_of_stations(origin_complex_id) # number of platforms at complex
onboardings = 0
destination_df = pd.DataFrame(columns=df_.columns)

# If there is only one station, we can just divide the ridership by the number of lines that go through that station
if num_stations == 1:
    num_lines = mta.lines_at_complex_id(origin_complex_id)
    onboardings += sum(df_["estimated_average_ridership"]) / len(num_lines)
    destination_df = df_ 
    destination_df["estimated_average_ridership"] = destination_df["estimated_average_ridership"] / len(num_lines)

# If there is more than one station
elif num_stations > 1:
    # For each destination station
    for i in range(len(df_)):
        # Get the lines that directly connect the origin and destination station
        destination_complex_id = str(df_.iloc[i]["destination_station_complex_id"])
        connecting_lines = mta.connecting_lines(origin_complex_id, destination_complex_id)

        # If there are lines that directly connect the origin and destination station
        if len(connecting_lines) > 0:
            # And if the line we are looking for is one of the connecting lines
            if line in connecting_lines:
                onboardings = onboardings + (df_.iloc[i]["estimated_average_ridership"] / len(connecting_lines))
                
                df_tmp = df_.iloc[[i]].copy()
                df_tmp["estimated_average_ridership"] /= len(connecting_lines)
                destination_df =  pd.concat([destination_df, df_tmp], ignore_index=True)
        # If there are no lines that directly connect the origin and destination station
        else:
            # Get all the lines that connect the origin and destination station
            lines = set()
            for path in mta.all_shortest_paths(origin_complex_id, "610"):
                connecting_lines = mta.connecting_lines(path[0], path[1])
                lines.update(connecting_lines)
            
            if line in lines:
                onboardings = onboardings + df_.iloc[i]["estimated_average_ridership"] / len(lines)

                df_tmp = df_.iloc[[i]].copy()
                df_tmp["estimated_average_ridership"] /= len(lines)
                destination_df =  pd.concat([destination_df, df_tmp], ignore_index=True)

print(onboardings)
destination_df
```

```python
def get_onboardings(origin_complex_id, line, df_):
    # Initialize variables
    num_stations = complexes.get_number_of_stations(origin_complex_id) # number of platforms at complex
    onboardings = 0
    destination_df = pd.DataFrame(columns=df_.columns)

    # If there is only one station, we can just divide the ridership by the number of lines that go through that station
    if num_stations == 1:
        num_lines = mta.lines_at_complex_id(origin_complex_id)
        onboardings += sum(df_["estimated_average_ridership"]) / len(num_lines)
        destination_df = df_ 
        destination_df["estimated_average_ridership"] = destination_df["estimated_average_ridership"] / len(num_lines)

    # If there is more than one station
    elif num_stations > 1:
        # For each destination station
        for i in range(len(df_)):
            # Get the lines that directly connect the origin and destination station
            destination_complex_id = str(df_.iloc[i]["destination_station_complex_id"])
            connecting_lines = mta.connecting_lines(origin_complex_id, destination_complex_id)

            # If there are lines that directly connect the origin and destination station
            if len(connecting_lines) > 0:
                # And if the line we are looking for is one of the connecting lines
                if line in connecting_lines:
                    onboardings = onboardings + (df_.iloc[i]["estimated_average_ridership"] / len(connecting_lines))

                    df_tmp = df_.iloc[[i]].copy()
                    df_tmp["estimated_average_ridership"] /= len(connecting_lines)
                    destination_df =  pd.concat([destination_df, df_tmp], ignore_index=True)
            # If there are no lines that directly connect the origin and destination station
            else:
                # Get all the lines that connect the origin and destination station
                lines = set()
                for path in mta.all_shortest_paths(origin_complex_id, "610"):
                    connecting_lines = mta.connecting_lines(path[0], path[1])
                    lines.update(connecting_lines)

                if line in lines:
                    onboardings = onboardings + df_.iloc[i]["estimated_average_ridership"] / len(lines)

                    df_tmp = df_.iloc[[i]].copy()
                    df_tmp["estimated_average_ridership"] /= len(lines)
                    destination_df =  pd.concat([destination_df, df_tmp], ignore_index=True)
    
    return onboardings, destination_df
```
